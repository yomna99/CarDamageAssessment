```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Récupérer l'image originale
image_path = '/content/istockphoto-175195079-612x612.jpg'
image = cv2.imread(image_path)

# Récupérer les résultats (assurez-vous que results est une liste)
for result in results:
    boxes = result.boxes.xyxy[0].cpu().numpy()  # Assurez-vous d'ajuster cela en fonction de votre structure de résultats
    masks = result.masks[0].cpu().numpy() if result.masks is not None else None  # Assurez-vous d'ajuster cela également

    # Récupérer les noms des classes
    class_names = list(result.names.values())

    # Dessiner les boîtes et les masques sur l'image
    for box, mask in zip(boxes, masks):
        x_min, y_min, x_max, y_max = map(int, box[:4])
        class_id = int(box[4])
        class_name = class_names[class_id]

        # Dessiner la boîte délimitatrice
        cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

        # Dessiner le masque
        color_mask = np.random.randint(0, 256, (1, 3), dtype=np.uint8)
        mask = (mask > 0.5).astype(np.uint8)  # Threshold pour obtenir une image binaire
        contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        cv2.drawContours(image, contours, -1, color_mask.tolist()[0], 2)

        # Ajouter le nom de la classe
        label = f"{class_name}: {box[5]:.2f}"
        cv2.putText(image, label, (x_min, y_min - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

# Afficher l'image résultante
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
